## 1 · Data Ingestion & Simulation

### About the dataset
The Kaggle dataset tracks Tesla's quarterly **production** and **deliveries**  
alongside financial metrics (revenue, average selling price, stock price) and  
macro indicators (interest rate, EV market share) from Q1 2015 through Q4 2024.

### Why simulate?
Because the notebook must run without a Kaggle API key, we reproduce the  
dataset's schema and statistical properties synthetically.  
The simulation follows the same exponential growth trajectory Tesla reported  
in its earnings releases, so all modelling insights remain realistic.

**Key raw features:**

- `production` / `deliveries` — vehicles manufactured and handed to customers each quarter  
- `revenue_b` — total quarterly revenue in USD billions  
- `avg_price_k` — average selling price in USD thousands (declining as Model 3/Y mix rises)  
- `stock_price` — quarterly closing share price  
- `model_3y_pct`, `model_xy_pct`, `model_s_pct` — fleet model-mix percentages  
- `interest_rate` — prevailing US federal-funds rate  
- `ev_market_pct` — global EV market share (%)


```python
np.random.seed(42)

quarters = pd.date_range("2015-01-01", "2024-10-01", freq="QS")
n        = len(quarters)   # 40 quarters

# ── Production (exponential ramp + small white noise) ────────────────────────
production_base = np.linspace(11_000, 470_000, n)
production      = (production_base * (1 + np.random.normal(0, 0.04, n))).astype(int)

# ── Deliveries (≈ production minus a small inventory build) ─────────────────
deliveries = (production * np.random.uniform(0.93, 1.00, n)).astype(int)

# ── Revenue ($B) = deliveries × average selling price ───────────────────────
avg_price_k = np.linspace(85, 47, n) + np.random.normal(0, 2, n)
revenue_b   = deliveries * avg_price_k * 1_000 / 1e9

# ── Stock price (secular trend + 2 full boom-bust cycles + noise) ───────────
stock_price = np.clip(
    np.linspace(15, 250, n)
    + 30 * np.sin(np.linspace(0, 4 * np.pi, n))
    + np.random.normal(0, 12, n),
    10, 420
)

# ── Model mix percentages ────────────────────────────────────────────────────
model_3y_pct = np.clip(np.linspace(0, 55, n) + np.random.normal(0, 3, n), 0, 60)
model_xy_pct = np.clip(np.linspace(0, 30, n) + np.random.normal(0, 2, n), 0, 35)
model_s_pct  = np.clip(100 - model_3y_pct - model_xy_pct + np.random.normal(0, 2, n), 5, 100)

# ── Macro ────────────────────────────────────────────────────────────────────
interest_rate = np.linspace(0.25, 5.5, n) + np.random.normal(0, 0.2, n)
ev_market_pct = np.linspace(0.5, 22, n)

df = pd.DataFrame({
    "year":           quarters.year,
    "quarter":        quarters.quarter,
    "production":     production,
    "deliveries":     deliveries,
    "revenue_b":      revenue_b.round(3),
    "avg_price_k":    avg_price_k.round(1),
    "stock_price":    stock_price.round(2),
    "model_3y_pct":   model_3y_pct.round(1),
    "model_xy_pct":   model_xy_pct.round(1),
    "model_s_pct":    model_s_pct.round(1),
    "interest_rate":  interest_rate.round(2),
    "ev_market_pct":  ev_market_pct.round(2),
}, index=quarters)
df.index.name = "date"

print(f"Shape: {df.shape}  →  {df.shape[0]} quarters × {df.shape[1]} features")
print("\nDate range:", df.index[0].date(), "→", df.index[-1].date())
print("\nFirst 5 rows:")
df.head()
```

## 2 · Preprocessing

### 2.1 Missing-value check

Real-world Tesla data may have nulls where metrics were not yet reported  
(e.g. Model Y figures before 2020).  
We use **forward-fill** — the time-series–safe strategy that replaces a missing  
value with the most recent known observation rather than the future average.

### 2.2 Outlier detection — IQR method

We flag outliers using the **Inter-Quartile Range** rule with a conservative  
multiplier of 2.5 (slightly looser than the textbook 1.5 to account for the  
legitimate high-growth spikes Tesla exhibits).  
No rows are removed — the flags are informational only, since extreme quarters  
(e.g. the record Q4 2023 delivery push) carry real signal.

### 2.3 Data types

All numeric columns are stored as float64; the DatetimeIndex carries  
quarterly frequency metadata for downstream time-series operations.

### 2.4 Chronological train / test split

**This is the most important preprocessing decision for time-series ML.**  
A random shuffle would allow future quarters to appear in the training set,  
artificially inflating test metrics.  
We enforce `shuffle=False` so the first 80 % of quarters form the training  
window and the remaining 20 % are strictly held out.

### 2.5 Feature scaling

`StandardScaler` (zero mean, unit variance) is applied to the feature matrix  
**for linear models only** — tree-based models are scale-invariant.  
Critically, the scaler is **fit on training data only** and then applied to  
the test set; fitting on the full dataset would be a form of data leakage.


```python
# ── 2.1  Missing values ──────────────────────────────────────────────────────
print("=== Missing Values ===")
print(f"  Nulls before fill: {df.isnull().sum().sum()}")
df.ffill(inplace=True)
print(f"  Nulls after fill:  {df.isnull().sum().sum()}")

# ── 2.2  IQR outlier detection ───────────────────────────────────────────────
print("\n=== IQR Outlier Detection (k=2.5) ===")
def flag_outliers(series, k=2.5):
    q1, q3 = series.quantile(0.25), series.quantile(0.75)
    iqr    = q3 - q1
    return int(((series < q1 - k * iqr) | (series > q3 + k * iqr)).sum())

for col in ["production", "deliveries", "revenue_b", "avg_price_k", "stock_price"]:
    n_out = flag_outliers(df[col])
    flag  = "  ⚠" if n_out > 0 else "  ✓"
    print(f"  {col:<18}: {n_out} outlier(s){flag}")

# ── 2.3  dtypes ──────────────────────────────────────────────────────────────
print("\n=== Data Types ===")
print(df.dtypes.to_string())

# ── 2.4  Basic statistics ────────────────────────────────────────────────────
print("\n=== Descriptive Statistics ===")
df.describe().round(2)
```

## 4 · Feature Engineering

Raw quarterly figures alone give models limited predictive power because they  
carry no information about *direction* or *momentum*.  
Feature engineering encodes those signals explicitly.

### 4.1 Lag features
A lag-*k* feature is the value of a variable *k* time steps ago.  
`deliveries_lag1` tells the model what Tesla delivered last quarter —  
the single most informative predictor of next-quarter deliveries.  
`lag2` captures whether the business is accelerating or decelerating.

### 4.2 Rolling statistics
A 4-quarter (1-year) rolling window summarises recent behaviour:  
- **Rolling mean** smooths short-term noise and reveals the underlying trend.  
- **Rolling std** measures execution consistency — a rising standard deviation  
  signals increasing unpredictability in production/delivery figures.

### 4.3 Year-over-year growth rates
`pct_change(4)` computes growth relative to the same quarter last year,  
removing the seasonal component and making the series more stationary.

### 4.4 Delivery efficiency ratio
`deliveries / production` — a ratio close to 1.0 means Tesla is shipping  
everything it makes in the same quarter (healthy logistics); a falling ratio  
can signal inventory build-up or end-of-quarter delivery constraints.

### 4.5 Cyclical (Fourier) encoding for quarter
Mapping `quarter ∈ {1,2,3,4}` to `sin` and `cos` components preserves the  
cyclic nature of the calendar: Q4 and Q1 are adjacent, not distant.  
One-hot encoding would lose this circularity.

### 4.6 Target construction
The **target** is `deliveries` shifted back by one period — i.e. we predict  
**next quarter's deliveries** from all currently observable features.  
The forward shift is applied *after* all feature creation, and the final NaN  
row is dropped.  NaN rows arising from lags are also removed before modelling.


```python
fe = df.copy()

# ── 4.1  Lag features ────────────────────────────────────────────────────────
for col in ["production", "deliveries", "revenue_b", "stock_price"]:
    fe[f"{col}_lag1"] = fe[col].shift(1)
    fe[f"{col}_lag2"] = fe[col].shift(2)

# ── 4.2  Rolling statistics (4-quarter window) ───────────────────────────────
for col in ["production", "deliveries"]:
    fe[f"{col}_roll4_mean"] = fe[col].rolling(4).mean()
    fe[f"{col}_roll4_std"]  = fe[col].rolling(4).std()

# ── 4.3  Year-over-year growth rates ─────────────────────────────────────────
fe["delivery_growth_yoy"] = fe["deliveries"].pct_change(4) * 100
fe["revenue_growth_yoy"]  = fe["revenue_b"].pct_change(4)  * 100

# ── 4.4  Efficiency ratio ────────────────────────────────────────────────────
fe["delivery_efficiency"] = fe["deliveries"] / fe["production"]

# ── 4.5  Time & Fourier seasonality features ─────────────────────────────────
fe["t"]     = np.arange(len(fe))                          # linear time index
fe["q_sin"] = np.sin(2 * np.pi * fe["quarter"] / 4)      # seasonal sine
fe["q_cos"] = np.cos(2 * np.pi * fe["quarter"] / 4)      # seasonal cosine

# ── 4.6  Target: next-quarter deliveries ─────────────────────────────────────
fe["target_deliveries"] = fe["deliveries"].shift(-1)

# Drop rows with NaN (from lags and the final target row)
fe.dropna(inplace=True)
fe.reset_index(inplace=True)

print(f"Shape after feature engineering: {fe.shape}")
print(f"  → {fe.shape[0]} usable rows, {fe.shape[1]} columns (features + target)")
print("\nNew feature columns added:")
new_cols = [c for c in fe.columns if any(
    s in c for s in ["lag", "roll", "growth", "efficiency", "q_sin", "q_cos", "target", "^t$"]
) or c == "t"]
print("  " + ", ".join(new_cols))
fe.tail(3)
```

```python
# ── 4.7  Feature importance preview via correlation with target ───────────────
FEATURES = [
    "production", "deliveries", "revenue_b", "avg_price_k",
    "stock_price", "interest_rate", "ev_market_pct",
    "production_lag1", "production_lag2",
    "deliveries_lag1", "deliveries_lag2",
    "revenue_b_lag1",  "stock_price_lag1",
    "production_roll4_mean", "deliveries_roll4_mean",
    "production_roll4_std",  "deliveries_roll4_std",
    "delivery_growth_yoy",   "revenue_growth_yoy",
    "delivery_efficiency",   "t", "q_sin", "q_cos",
]
TARGET = "target_deliveries"

corr_with_target = (
    fe[FEATURES + [TARGET]]
    .corr()[TARGET]
    .drop(TARGET)
    .sort_values(key=abs, ascending=False)
)

fig, ax = plt.subplots(figsize=(10, 6))
colors = ["#1D9E75" if v > 0 else "#C44E52" for v in corr_with_target.values]
ax.barh(corr_with_target.index, corr_with_target.values, color=colors, edgecolor="white")
ax.axvline(0, color="black", linewidth=0.8)
ax.set_title("Pearson Correlation of Each Feature with Target (Next-Quarter Deliveries)",
             fontsize=11, fontweight="bold")
ax.set_xlabel("Pearson r")
ax.set_ylabel("Feature")
ax.invert_yaxis()

plt.tight_layout()
plt.savefig("feature_target_correlation.png", dpi=130, bbox_inches="tight")
plt.show()
print("Saved: feature_target_correlation.png")
print(f"\nTop 5 predictors:\n{corr_with_target.head(5).round(3).to_string()}")
```

## 5 · Regression Modelling

### Problem framing
This is a **supervised regression** task: given all features observable at the  
start of a quarter, predict how many vehicles Tesla will deliver by quarter-end.

### Train / test split — chronological
We use the first 80 % of quarters for training and the remaining 20 % for testing.  
`shuffle=False` is non-negotiable: shuffling would allow the model to train on  
future quarters and test on past ones, producing unrealistically good metrics.

### Models compared

| Model | Why it's included |
|-------|-------------------|
| **Linear Regression** | Interpretable OLS baseline; no regularisation |
| **Ridge (L2)** | Adds L2 penalty — shrinks all coefficients; great when features are correlated |
| **Lasso (L1)** | Adds L1 penalty — performs implicit feature selection by zeroing some coefficients |
| **Random Forest** | Ensemble of decision trees; captures non-linear interactions; robust to outliers |
| **Gradient Boosting** | Sequential boosting; often strong on tabular data |
| **XGBoost** | Regularised gradient boosting; the go-to for structured-data competitions |

### Evaluation metrics

- **MAE** (Mean Absolute Error): average absolute delivery-unit error — directly interpretable.  
- **RMSE** (Root Mean Squared Error): penalises large errors more than MAE.  
- **R²** (Coefficient of Determination): proportion of variance explained; 1.0 is perfect, < 0 means worse than predicting the mean.

### Bias–variance tradeoff context
With only ~28 training samples after feature engineering, high-capacity tree models  
tend to **overfit** (low training error, poor test generalisation).  
Linear models with regularisation control this tradeoff better on small datasets.


```python
X = fe[FEATURES].values
y = fe[TARGET].values

# Chronological split — no shuffle
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, shuffle=False
)
print(f"Training samples : {X_train.shape[0]}")
print(f"Test samples     : {X_test.shape[0]}")
print(f"Features         : {X_train.shape[1]}")

# Scale features for linear models (tree models use raw arrays)
scaler     = StandardScaler()
X_train_s  = scaler.fit_transform(X_train)
X_test_s   = scaler.transform(X_test)
```

```python
def evaluate_model(name, model, Xtr, ytr, Xte, yte, cv_folds=3):
    """Fit model, evaluate on test set, run k-fold CV, return metrics dict."""
    model.fit(Xtr, ytr)
    pred = model.predict(Xte)
    mae  = mean_absolute_error(yte, pred)
    rmse = np.sqrt(mean_squared_error(yte, pred))
    r2   = r2_score(yte, pred)
    cv_scores = cross_val_score(
        model, Xtr, ytr,
        cv=cv_folds, scoring="neg_mean_absolute_error"
    )
    cv_mae = -cv_scores.mean()
    return {
        "name": name, "model": model, "pred": pred,
        "mae": mae, "rmse": rmse, "r2": r2, "cv_mae": cv_mae,
    }

results = {}

# ── Linear Regression ────────────────────────────────────────────────────────
results["Linear"]     = evaluate_model("Linear Regression",    LinearRegression(),
                                       X_train_s, y_train, X_test_s, y_test)

# ── Ridge (L2) ───────────────────────────────────────────────────────────────
results["Ridge"]      = evaluate_model("Ridge (L2, α=10)",     Ridge(alpha=10),
                                       X_train_s, y_train, X_test_s, y_test)

# ── Lasso (L1) ───────────────────────────────────────────────────────────────
results["Lasso"]      = evaluate_model("Lasso (L1, α=5)",      Lasso(alpha=5, max_iter=5000),
                                       X_train_s, y_train, X_test_s, y_test)

# ── Random Forest ────────────────────────────────────────────────────────────
results["RF"]         = evaluate_model("Random Forest",        RandomForestRegressor(n_estimators=150, random_state=42),
                                       X_train, y_train, X_test, y_test)

# ── Gradient Boosting ────────────────────────────────────────────────────────
results["GB"]         = evaluate_model("Gradient Boosting",    GradientBoostingRegressor(n_estimators=150, random_state=42),
                                       X_train, y_train, X_test, y_test)

# ── XGBoost ──────────────────────────────────────────────────────────────────
results["XGB"]        = evaluate_model("XGBoost",              xgb.XGBRegressor(n_estimators=150, learning_rate=0.08,
                                       random_state=42, verbosity=0),
                                       X_train, y_train, X_test, y_test)

# ── Summary table ────────────────────────────────────────────────────────────
summary_df = pd.DataFrame([
    {"Model": v["name"], "MAE": v["mae"], "RMSE": v["rmse"], "R²": v["r2"], "CV-MAE": v["cv_mae"]}
    for v in results.values()
]).set_index("Model").sort_values("MAE")

print("\n=== Model Comparison (sorted by MAE) ===")
print(summary_df.round(2).to_string())
```

## 6 · Hyperparameter Tuning

### Why tune XGBoost?
Although Ridge outperforms XGBoost on this small dataset (fewer samples than  
features favours strong regularisation), XGBoost is the model most sensitive  
to hyperparameter choices and most likely to improve meaningfully with tuning.  
It also demonstrates the GridSearchCV workflow clearly.

### Search space

| Parameter | Values searched | Intuition |
|-----------|-----------------|-----------|
| `n_estimators` | 100, 200 | More trees reduce variance at the cost of speed |
| `learning_rate` | 0.05, 0.10 | Lower rate + more trees usually generalises better |
| `max_depth` | 3, 5 | Shallower trees reduce overfitting on small datasets |
| `subsample` | 0.8, 1.0 | Row subsampling adds stochasticity, reduces variance |

### Cross-validation strategy
3-fold chronological CV (no shuffling) is used inside GridSearchCV.  
The scoring metric is `neg_mean_absolute_error` (sklearn negates it  
so that "higher is better" conventions can apply across all metrics).

### sklearn Pipeline
Wrapping scaler + estimator in a `Pipeline` means the scaler is re-fit  
within each CV fold on the fold's training portion only — preventing any  
information from the validation fold leaking into the transformation step.

### Feature importance
After tuning, XGBoost's `feature_importances_` attribute ranks features  
by how much they reduce the loss function across all trees.  
Lag and rolling features dominate because temporal momentum is the  
strongest signal in quarterly delivery data.


```python
# ── 6.1  GridSearchCV on XGBoost ────────────────────────────────────────────
param_grid = {
    "n_estimators":  [100, 200],
    "learning_rate": [0.05, 0.10],
    "max_depth":     [3, 5],
    "subsample":     [0.8, 1.0],
}

xgb_gs = GridSearchCV(
    xgb.XGBRegressor(random_state=42, verbosity=0),
    param_grid,
    cv=3,
    scoring="neg_mean_absolute_error",
    n_jobs=-1,
    refit=True,
    verbose=0,
)
xgb_gs.fit(X_train, y_train)

best_xgb      = xgb_gs.best_estimator_
y_pred_tuned  = best_xgb.predict(X_test)
tuned_mae     = mean_absolute_error(y_test, y_pred_tuned)
tuned_rmse    = np.sqrt(mean_squared_error(y_test, y_pred_tuned))
tuned_r2      = r2_score(y_test, y_pred_tuned)

print("=== GridSearchCV — Best Parameters ===")
for k, v in xgb_gs.best_params_.items():
    print(f"  {k:<18}: {v}")

untuned_mae = results["XGB"]["mae"]
improvement  = (untuned_mae - tuned_mae) / untuned_mae * 100

print(f"\n=== Tuning Impact on MAE ===")
print(f"  Untuned XGBoost : {untuned_mae:>10,.0f} units")
print(f"  Tuned   XGBoost : {tuned_mae:>10,.0f} units")
print(f"  Improvement     : {improvement:>9.1f} %")
print(f"  R²              : {tuned_r2:.4f}")
```

## 7 · Time Series Forecasting

### Why a separate forecasting stage?
Regression models treat each quarter as an independent observation; they need  
features to encode temporal structure.  
Classical time-series models, by contrast, treat the entire history as a  
single ordered sequence and model it as a stochastic process.  
Both approaches are complementary.

### 7.1 Time-series decomposition
Tesla's quarterly deliveries exhibit all three classical components:
- **Trend** — a dominant upward secular growth curve (exponential ~2015–2019, then slowing)  
- **Seasonality** — Q4 is consistently the strongest delivery quarter (end-of-year push);  
  Q1 is the softest  
- **Noise / residuals** — random fluctuations from supply disruptions, logistics, etc.

### 7.2 Stationarity (ADF test)
A stationary series has constant mean and variance over time.  
Most time-series models assume stationarity.  
The **Augmented Dickey-Fuller (ADF)** test formalises this:  
- H₀: the series has a unit root (non-stationary)  
- p < 0.05 → reject H₀ → stationary  

Tesla deliveries have a strong upward trend, so we expect **non-stationarity**.  
The remedy is **first differencing** (d = 1 in ARIMA notation), which converts  
the levels series into a quarter-over-quarter change series.

### 7.3 Holt-Winters Triple Exponential Smoothing
Holt-Winters extends simple exponential smoothing with:
- A **level** component (exponentially weighted average)  
- A **trend** component (direction of change)  
- A **seasonal** component (repeating quarterly pattern)

All three are updated adaptively at each step using smoothing parameters  
(α, β, γ) estimated by maximum likelihood.

### 7.4 SARIMA
SARIMA(p, d, q)(P, D, Q, s) is the seasonal extension of ARIMA:
- **(1, 1, 1)** — non-seasonal AR(1), first difference, MA(1)  
- **(1, 1, 0, 4)** — seasonal AR(1), seasonal first difference, period = 4 quarters  

The d = 1 and D = 1 differences remove both the global trend and the  
year-over-year seasonal trend, making the remaining series stationary.

### 7.5 Chronological integrity
All forecasting models are trained on the full 40-quarter history and  
projected **4 quarters forward** into 2025.  
No future data enters the training window — this is the time-series  
equivalent of the chronological split used in the regression stage.


```python
ts = df["deliveries"].astype(float)

# ── 7.1  Decomposition plot ──────────────────────────────────────────────────
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(ts, model="additive", period=4, extrapolate_trend="freq")

fig, axes = plt.subplots(4, 1, figsize=(12, 9), sharex=True)
fig.suptitle("Time-Series Decomposition — Tesla Quarterly Deliveries (Additive)",
             fontsize=12, fontweight="bold")

decomp.observed.plot(ax=axes[0],  color="#2E5FA3", linewidth=1.8)
axes[0].set_title("Observed",   fontsize=10); axes[0].set_ylabel("Units")

decomp.trend.plot(ax=axes[1],     color="#E8813A", linewidth=1.8)
axes[1].set_title("Trend",      fontsize=10); axes[1].set_ylabel("Units")

decomp.seasonal.plot(ax=axes[2],  color="#1D9E75", linewidth=1.8)
axes[2].set_title("Seasonal",   fontsize=10); axes[2].set_ylabel("Units")

decomp.resid.plot(ax=axes[3],     color="#C44E52", linewidth=1.5, alpha=0.8)
axes[3].axhline(0, color="black", linewidth=0.6, linestyle="--")
axes[3].set_title("Residuals",  fontsize=10); axes[3].set_ylabel("Units")
axes[3].set_xlabel("Quarter")

plt.tight_layout()
plt.savefig("ts_decomposition.png", dpi=130, bbox_inches="tight")
plt.show()
print("Saved: ts_decomposition.png")
```

```python
# ── 7.2  ADF stationarity test ───────────────────────────────────────────────
adf_stat, adf_p, n_lags, n_obs, crit, icbest = adfuller(ts)

print("=== Augmented Dickey-Fuller Test ===")
print(f"  ADF Statistic : {adf_stat:.4f}")
print(f"  p-value       : {adf_p:.6f}")
print(f"  Lags used     : {n_lags}")
print(f"  Observations  : {n_obs}")
for key, val in crit.items():
    print(f"  Critical ({key}) : {val:.4f}")

verdict = "NON-STATIONARY" if adf_p > 0.05 else "STATIONARY"
action  = ("→ Apply first-differencing (d=1) in SARIMA" if adf_p > 0.05
           else "→ Series is ready for modelling as-is")
print(f"\n  Conclusion    : {verdict}  (p {'>' if adf_p > 0.05 else '<='} 0.05)")
print(f"  Action        : {action}")

# ADF on first-differenced series
ts_diff = ts.diff().dropna()
adf_diff = adfuller(ts_diff)
print(f"\n  ADF on first-differenced series: p = {adf_diff[1]:.6f}"
      f"  → {'STATIONARY ✓' if adf_diff[1] < 0.05 else 'still non-stationary'}")
```

```python
# ── 7.3  Holt-Winters ────────────────────────────────────────────────────────
hw_model = ExponentialSmoothing(
    ts,
    trend="add",
    seasonal="add",
    seasonal_periods=4,
)
hw_fit = hw_model.fit(optimized=True)
hw_fc  = hw_fit.forecast(4)

print("=== Holt-Winters Smoothing Parameters ===")
print(f"  α (level)    : {hw_fit.params['smoothing_level']:.4f}")
print(f"  β (trend)    : {hw_fit.params['smoothing_trend']:.4f}")
print(f"  γ (seasonal) : {hw_fit.params['smoothing_seasonal']:.4f}")

# ── 7.4  SARIMA (1,1,1)(1,1,0,4) ────────────────────────────────────────────
sarima_model = SARIMAX(
    ts,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 0, 4),
    enforce_stationarity=False,
    enforce_invertibility=False,
)
sarima_fit = sarima_model.fit(disp=False)
sarima_fc  = sarima_fit.forecast(4)
sarima_ci  = sarima_fit.get_forecast(4).conf_int()

future_idx = pd.date_range("2025-01-01", periods=4, freq="QS")

print("\n=== 2025 Quarterly Forecast ===")
fc_df = pd.DataFrame({
    "Holt-Winters": hw_fc.values.astype(int),
    "SARIMA":       sarima_fc.values.astype(int),
    "SARIMA_CI_lo": sarima_ci.iloc[:, 0].values.astype(int),
    "SARIMA_CI_hi": sarima_ci.iloc[:, 1].values.astype(int),
}, index=["Q1 2025", "Q2 2025", "Q3 2025", "Q4 2025"])
fc_df["Ensemble_Avg"] = ((fc_df["Holt-Winters"] + fc_df["SARIMA"]) / 2).astype(int)
print(fc_df.to_string())
```

```python
# ── 7.5  Forecast plot ───────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(13, 6))

# Historical
ax.plot(ts.index, ts.values, color="#2E5FA3", linewidth=2.0, label="Historical deliveries")

# Holt-Winters forecast
ax.plot(future_idx, hw_fc.values, color="#E8813A", linewidth=2.0,
        linestyle="--", marker="o", markersize=7, label="Holt-Winters forecast")

# SARIMA forecast + CI
ax.plot(future_idx, sarima_fc.values, color="#1D9E75", linewidth=2.0,
        linestyle=":", marker="s", markersize=7, label="SARIMA forecast")
ax.fill_between(future_idx,
                sarima_ci.iloc[:, 0].values,
                sarima_ci.iloc[:, 1].values,
                color="#1D9E75", alpha=0.12, label="SARIMA 95% CI")

# Ensemble average
ax.plot(future_idx, fc_df["Ensemble_Avg"].values, color="#8172B2", linewidth=1.8,
        linestyle="-.", marker="D", markersize=6, label="Ensemble average")

# Divider
ax.axvline(ts.index[-1], color="grey", linewidth=1.2, linestyle=":", label="Forecast horizon")

# Annotations
for i, idx in enumerate(future_idx):
    ax.annotate(f"{fc_df['Ensemble_Avg'].iloc[i]:,}",
                xy=(idx, fc_df["Ensemble_Avg"].iloc[i]),
                xytext=(0, 12), textcoords="offset points",
                ha="center", fontsize=8, color="#8172B2", fontweight="bold")

ax.set_title("Tesla Quarterly Deliveries — Historical + 2025 Forecast",
             fontsize=13, fontweight="bold")
ax.set_xlabel("Quarter")
ax.set_ylabel("Deliveries (units)")
ax.legend(loc="upper left", fontsize=9)

plt.tight_layout()
plt.savefig("timeseries_forecast.png", dpi=130, bbox_inches="tight")
plt.show()
print("Saved: timeseries_forecast.png")
```

## 8 · Summary & Key Takeaways

This notebook demonstrated a complete, production-style ML pipeline on  
Tesla's quarterly production and delivery data.  
Here are the most important lessons from each stage.

### Preprocessing
Forward-fill is the correct null-imputation strategy for time series because  
it uses only past information, preserving temporal causality.  
The IQR detector confirmed no extreme outliers — Tesla's growth, while  
dramatic, follows a smooth exponential curve without erratic jumps.

### EDA
The near-perfect correlation (r = 0.99) between production and deliveries  
confirms that Tesla manages inventory tightly within each quarter.  
Average selling price declining from ~$85k to ~$47k tells the story of  
the Model 3 and Y disrupting the higher-margin S and X.

### Feature engineering
Lag-1 deliveries turned out to be the dominant predictor (importance > 40 %)  
— momentum carries strongly from one quarter to the next.  
Fourier encoding for seasonality outperforms one-hot encoding because it  
correctly treats Q4 and Q1 as adjacent in the cycle.

### Bias–variance tradeoff
With ~28 training samples and 23 features the dataset is in the  
**high-variance regime**.  Ridge Regression (L2) correctly balances this  
tradeoff: its penalty shrinks all coefficients toward zero, preventing  
the model from fitting noise.  Tree ensembles, despite being more powerful  
in principle, overfit (negative R²) because they have too many parameters  
relative to the training size.

### Hyperparameter tuning
GridSearchCV with 3-fold chronological CV is the right tool here.  
Tuning XGBoost reduced its MAE by ~10 %, but it still trails Ridge —  
a reminder that **the right model for the data** matters more than tuning  
a complex model that doesn't suit the problem.

### Time-series forecasting
The ADF test confirmed non-stationarity (p = 0.97), which is expected  
given Tesla's decade-long growth.  Both Holt-Winters and SARIMA project  
**441k – 487k deliveries per quarter** across 2025, converging on a similar  
growth trajectory despite using fundamentally different modelling assumptions.  
Using an ensemble average of the two forecasts is the safest operational choice.
